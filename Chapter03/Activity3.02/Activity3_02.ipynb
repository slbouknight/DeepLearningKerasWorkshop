```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import random
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
X = pd.read_csv('../data/HCV_feats.csv')
y = pd.read_csv('../data/HCV_target.csv')
```

```python
print(f'Number of examples in dataset: {X.shape[0]}')
print(f'Number of features: {X.shape[1]}')
print("Possible Output Classes = ", y['AdvancedFibrosis'].unique())
```

```python
seed = 1
np.random.seed(seed)
random.set_seed(seed)
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns = X.columns)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
print(f'Number of examples in training set = {X_train.shape[0]}')
print(f'Number of examples in test set = {X_test.shape[0]}')
```

```python
np.random.seed(seed)
random.set_seed(seed)

classifier = Sequential()
classifier.add(Dense(3, activation='tanh', input_dim=X_train.shape[1]))
classifier.add(Dense(1, activation='sigmoid'))
classifier.compile(optimizer='sgd', loss='binary_crossentropy', metrics='accuracy')

classifier.summary()
```

```python
history = classifier.fit(X_train, y_train, batch_size=20, epochs=100, validation_split=0.1, shuffle=False)
```

```python
matplotlib.rcParams['figure.figsize'] = (10.0, 9.0)

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'validation loss'], loc = 'upper right')
```

```python
print(f"Best Accuracy on training set = {max(history.history['accuracy'])*100:.3f}%")
print(f"Best Accuracy on validation set = {max(history.history['val_accuracy'])*100:.3f}%") 

test_loss, test_acc = classifier.evaluate(X_test, y_test['AdvancedFibrosis'])
print(f'The loss on the test set is {test_loss:.4f} and the accuracy is {test_acc*100:.3f}%')
```

```python
np.random.seed(seed)
random.set_seed(seed)

classifier = Sequential()
classifier.add(Dense(4, activation='tanh', input_dim=X_train.shape[1]))
classifier.add(Dense(2, activation='tanh'))
classifier.add(Dense(1, activation='sigmoid'))
classifier.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])

classifier.summary()
```

```python
history = classifier.fit(X_train, y_train, batch_size=20, epochs=100, validation_split=0.1, shuffle=False)
```

```python
matplotlib.rcParams['figure.figsize'] = (10.0, 8.0)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'validation loss'], loc = 'upper right')
```

```python
print(f"Best Accuracy on training set = {max(history.history['accuracy'])*100:.3f}%")
print(f"Best Accuracy on test set = {max(history.history['val_accuracy'])*100:.3f}%")

test_loss, test_acc = classifier.evaluate(X_test, y_test['AdvancedFibrosis'])
print(f'The loss on the test set is {test_loss:.4f} and the accuracy is {test_acc*100:.3f}%')
```
